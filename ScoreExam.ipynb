```python
# 성적처리용 프로그램을 만들어본다.
# 1. 키보드로 개인정보를 입력받는다.
# 2. 키보드로 성적을 입력받는다. (국어, 영어, 수학, 파이썬)
# 3. 입력된 정보를 확인한다. -> 잘못된 정보를 수정한다.
# 4. 통계를 만들어 출력한다.
# 5. 등급을 출력하고 종료한다.

print("="*80)
print("성적처리 프로그램입니다. 개인정보를 입력하세요.")
name = input("이름 : ")
age = int(input("나이 : "))
address = input("주소 : ")
print("="*80)

print(f"{name} 님 환영합니다.")
print("="*80)

score = {'국어':0, '영어':0, '수학':0, '파이썬':0}
print(f"{name} 님의 성적을 입력하세요.")
for i in score.keys():
    score[i] = int(input(f"{i}성적 : "))
print("="*80)

while True:
    print("성적을 확인합니다.")
    for i in score.keys():
        print(f"{i} : {score[i]}")
    print("="*80)
    que1 = int(input("입력하신 성적을 수정할까요? 수정:1, 유지:0 "))
    if que1 == 0:
        break
    else:
        print("성적 수정을 진행합니다.")
        for i in score.keys():
            score[i] = int(input(f"{i}성적 : "))
        print("="*80)
total = sum(score.values())
avg = total / 4
print("="*80)

que2 = int(input("총점과 평균을 출력할까요? 출력:1, 패스:0 "))
if que2 == 1:
    print(f"{name} 님의 총점은 {total}, 평균은 {avg} 입니다.")
print("="*80)

que3 = int(input(f"{name} 님의 등급을 출력할까요? 출력:1 패스:0 "))
if que3 == 1:
    if avg >= 90:
        grade = 'A'
    elif avg >= 80:
        grade = 'B'
    elif avg >= 70:
        grade = 'C'
    elif avg >= 60:
        grade = 'D'
    else:
        grade = 'F'
    print(f"{name} 님은 {grade} 등급입니다.")
print("="*80)

print("성적처리 프로그램을 종료합니다.")
```
