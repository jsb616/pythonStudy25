```python
# 성적처리 프로그램

# 여러 사람의 성적 관리를 위한 리스트 생성
name = []
kor = []
eng = []
mat = []
pyt = []
total = []
avg = []
grade = []

prompt = '''
    1. 성적 추가
    2. 성적 확인
    3. 성적 수정
    4. 성적 삭제
    5. 프로그램 종료
'''

print("성적처리 프로그램입니다. 아래 메뉴를 보고 번호를 선택해주세요.")

run = True
while run:
    print(prompt)
    select = int(input("번호 입력(1~5) : "))
    print("=" * 80)

    if select == 1:
        print("성적을 추가합니다.")
        name.append(input("이름 : "))
        kor.append(int(input("국어 성적 : ")))
        eng.append(int(input("영어 성적 : ")))
        mat.append(int(input("수학 성적 : ")))
        pyt.append(int(input("파이썬 성적 : ")))
        total.append(kor[-1] + eng[-1] + mat[-1] + pyt[-1])
        avg.append(total[-1] / 4)
        if avg[-1] >= 90:
            grade.append('A')
        elif avg[-1] >= 80:
            grade.append('B')
        elif avg[-1] >= 70:
            grade.append('C')
        elif avg[-1] >= 60:
            grade.append('D')
        else:
            grade.append('F')
        print("성적이 추가되었습니다.")

    elif select == 2:
        print("성적을 확인합니다.")
        x = input("성적을 확인할 사람의 이름을 입력하세요 : ")
        if x in name:   # 입력이 되어 있는 사람
            for i in range(len(name)):
                if name[i] == x:
                    print(f"{x} 님의 성적을 출력합니다.")
                    print(f"국어:{kor[i]} \t 영어:{eng[i]} \t 수학:{mat[i]} \t 파이썬:{pyt[i]}")
                    print(f"총점:{total[i]} \t 평균:{avg[i]} \t 등급:{grade[i]}")
        else:   # 입력이 안 되어 있는 사람
            print(f"{x} 님의 성적 정보를 찾을 수 없습니다.")

    elif select == 3:
        print("성적을 수정합니다.")
        x = input("성적을 수정할 사람의 이름을 입력하세요 : ")
        if x in name:   # 입력이 되어 있는 사람
            for i in range(len(name)):
                if name[i] == x:
                    print(f"{x} 님의 성적을 수정합니다.")
                    kor[i] = int(input("국어 성적 : "))
                    eng[i] = int(input("영어 성적 : "))
                    mat[i] = int(input("수학 성적 : "))
                    pyt[i] = int(input("파이썬 성적 : "))
                    total[i] = kor[i] + eng[i] + mat[i] + pyt[i]
                    avg[i] = total[i] / 4
                    if avg[i] >= 90:
                        grade[i] = 'A'
                    elif avg[i] >= 80:
                        grade[i] = 'B'
                    elif avg[i] >= 70:
                        grade[i] = 'C'
                    elif avg[i] >= 60:
                        grade[i] = 'D'
                    else:
                        grade[i] = 'F'
                    print("성적이 수정되었습니다.")
        else:   # 입력이 안 되어 있는 사람
            print(f"{x} 님의 성적 정보를 찾을 수 없습니다.")

    elif select == 4:
        print("성적을 삭제합니다.")
        x = input("성적을 삭제할 사람의 이름을 입력하세요 : ")
        if x in name:   # 입력이 되어 있는 사람
            for i in range(len(name)):
                if name[i] == x:
                    print(f"{x} 님의 성적 정보가 삭제되었습니다.")
                    del name[i]
                    del kor[i]
                    del eng[i]
                    del mat[i]
                    del pyt[i]
                    del total[i]
                    del avg[i]
                    del grade[i]
        else:   # 입력이 안 되어 있는 사람
            print(f"{x} 님의 성적 정보를 찾을 수 없습니다.")

    elif select == 5:
        print("프로그램을 종료합니다.")
        run = False

    else:
        print("잘못된 입력입니다.")
```

```python
students = []

print("=== 성적관리 시스템 ===")
while True:
  menu = input("1. 학생등록  2. 학생목록  3. 학생조회  4. 학생수정  5. 학생삭제  6. 종료 >> ")
  print("=" * 70)

  if menu == '1':
    print("학생을 등록합니다.")
    student = {}
    student['name'] = input("학생 이름 : ")
    student['kor'] = int( input("국어 점수 : ") )
    student['eng'] = int( input("영어 점수 : ") )
    student['math'] = int( input("수학 점수 : ") )
    student['python'] = int( input("파이썬 점수 : ") )
    student['total'] = student['kor'] + student['eng'] + student['math'] + student['python']
    student['avg'] = student['total'] / 4

    if student['avg'] >= 90:
      student['grade'] = 'A'
    elif student['avg'] >= 80:

      student['grade'] = 'B'
    elif student['avg'] >= 70:
      student['grade'] = 'C'
    elif student['avg'] >= 60:
      student['grade'] = 'D'
    else:
      student['grade'] = 'F'

    students.append(student)
    print("✔ 학생등록 완료")

  elif menu == '2':
    if not students:
      print("❌ 등록된 학생이 없습니다.")
      continue
    print("학생 전체 목록을 조회합니다.\n")
    print("번호\t이름\t국어\t영어\t수학\t파이썬\t합계\t평균\t등급")
    print("=" * 70)

    i = 0
    for student in students:
      i += 1
      print(f"{i}\t{student['name']}\t{student['kor']}\t{student['eng']}\t{student['math']}\t{student['python']}\t{student['total']}\t{student['avg']:.2f}\t{student['grade']}")
    print("=" * 70)

  elif menu == '3':
    no = int( input("조회할 학생 번호를 입력하세요 : ") )

    if no < 1 or no > len(students):
      print(f"{no}번 학생은 존재하지 않습니다.")
      continue

    print("번호\t이름\t국어\t영어\t수학\t파이썬\t합계\t평균\t등급")
    print("=" * 70)
    print(f"{no}\t{students[no-1]['name']}\t{students[no-1]['kor']}\t{students[no-1]['eng']}\t{students[no-1]['math']}\t{students[no-1]['python']}\t{students[no-1]['total']}\t{students[no-1]['avg']:.2f}\t{students[no-1]['grade']}")

  elif menu == '4':
    no = int( input("수정할 학생 번호를 입력하세요 : ") )

    if no < 1 or no > len(students):
      print(f"{no}번 학생은 존재하지 않습니다.")
      continue

    print("학생정보를 수정합니다.")
    student = {}
    student['name'] = input("학생 이름 : ")
    student['kor'] = int( input("국어 점수 : ") )
    student['eng'] = int( input("영어 점수 : ") )
    student['math'] = int( input("수학 점수 : ") )
    student['python'] = int( input("파이썬 점수 : ") )
    student['total'] = student['kor'] + student['eng'] + student['math'] + student['python']
    student['avg'] = student['total'] / 4

    if student['avg'] >= 90:
      student['grade'] = 'A'
    elif student['avg'] >= 80:
      student['grade'] = 'B'
    elif student['avg'] >= 70:
      student['grade'] = 'C'
    elif student['avg'] >= 60:
      student['grade'] = 'D'
    else:
      student['grade'] = 'F'

    students[no-1] = student
    print("✔ 학생정보 수정완료")

  elif menu == '5':
    no = int( input("삭제할 학생 번호를 입력하세요 : ") )

    if no < 1 or no > len(students):
      print(f"{no}번 학생은 존재하지 않습니다.")
      continue

    if input(f"{no}번 학생정보를 삭제하시겠습니까 ? (삭제 : Y, 취소 : N) >> ").lower() == 'y':
      students.remove(students[no-1])
      print(f"✔ {no}번 학생 삭제 완료")
  elif menu == '6':
    print("=== 프로그램을 종료합니다 ===")
    break

  print("\n\n")
```
