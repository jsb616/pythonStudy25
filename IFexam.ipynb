if문은 제어문으로 참, 거짓을 판단하여 흐름을 조절함

**if 문법**
```
if 변수(참/거짓):
    참일 때 실행

else:
    거짓일 때 실행

    
```
반드시 tab(탭)을 해줘야 함

````python
# ex: 만약에 돈이 있으면 택시를 타고 가고, 없으면 걸어가라
#     if          True  print("택시를 탐")  False  print("걸어감")

money = True                    # 돈이 있다.
if money:                       # 만약에 money값 안에 True이면 아래 들여쓰기한 곳을 실행한다.
    print("택시를 타고 가라")   # True일 때 출력문
else:                           # 위 if문에 False가 들어올 때 아래 들여쓰기한 곳을 실행한다.
    print("걸어가라")
print("학교에 도착!")
````

````python
# ex: 돈이 3000원 이상 있거나 카드가 있다면 택시를 타고 가고, 그렇지 않으면 걸어가라
method = input("돈이면 money나 m, 카드면 card나 c를 입력해주세요:")
if method in ['money', 'm']:
    money = int(input("돈이 얼마있나요? "))
    if money >= 3000:
        print("택시를 타고 가라")
    else:
        print("걸어가라")
elif method in ['card', 'c']:
    print("택시를 타고 가라")
else:
    print("잘못된 입력입니다.")
````

````python
# ex: 주머니에 돈이 있으면 택시를 타고 가고, 없으면 걸어가라
pocket = ['paper', 'cellphone', 'money']
if 'money' in pocket:
    print("택시를 타고 가라")
else:
    print("걸어가라")
````

````python
# ex: 주머니에 돈이 있으면 택시를 타고 가고, 주머니에 돈은 없지만 카드가 있으면 택시를 타고 가고, 돈도 없고 카드도 없으면 걸어가라
pocket = ['paper', 'cellphone', 'card']
if 'money' in pocket or 'card' in pocket:
    print("택시를 타고 가라")
else:
    print("걸어가라")
````

조건부 표현식

````python
score = int(input("점수를 입력하세요(0~100): "))

if score >= 60:
    message = "success"
else:
    message = "failure"
print(message)
````

````python
# 조건부 표현식을 사용하면 위 코드를 간단히 표현 가능
score = int(input("점수를 입력하세요: "))
message = "success" if score >= 60 else "failure"
print(message)
````

````python
age = int(input("나이를 입력하세요: "))
adult = "성인입" if age >= 18 else "성인이 아닙"
print(f"당신은 {adult}니다.")
````

````python
age = int(input("나이를 입력하세요: "))
if age >= 18:
    x = "성인"
elif age >= 12:
    x = "청소년"
else:
    x = "어린이/유아"
print(f"당신은 {x}입니다.")
````
